### What is Data Visualization ?

Data visualization is the practice of translating information into a visual context, such as a map or graph, to make data easier for the human brain to understand and pull insights from.

### Why We Need Data Visualization?

1) It in help to visual patterns, trends and correlations between features that might not otherwise be detected.

2) Data visualization helps to tell stories by curating data into a form easier to understand, highlighting the trends and outliers.

3) An increased understanding of the next steps that must be taken to improve the organization

4) An improved ability to maintain the audience's interest with information they can understand.

### 1D Scatter plot

```python
df["species"].unique()
```

```python
Iris_setosa = df.loc[df["species"] == "Iris-setosa"]
Iris_versicolor = df.loc[df["species"] == "Iris-versicolor"]
Iris_virginica = df.loc[df["species"] == "Iris-virginica"]
```

```python
sns.scatterplot(Iris_setosa["petal_width"],np.zeros_like(Iris_setosa["petal_width"]),data = Iris_setosa)
sns.scatterplot(Iris_versicolor["petal_width"],np.zeros_like(Iris_versicolor["petal_width"]),data = Iris_versicolor)
sns.scatterplot(Iris_virginica["petal_width"],np.zeros_like(Iris_virginica["petal_width"]),data = Iris_virginica)
plt.grid()
```

Green points are Virginica, yellow points are Versicolor and blue points are Setosa

### 2-D plot

1) petal_length V/S petal_width

```python
plt.figure(figsize=(16, 6))
sns.set_style("whitegrid")
sns.scatterplot(x="petal_length",y="petal_width",data=df,hue="species",style="species")
```

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
sns.scatterplot(x="sepal_length",y="sepal_width",data=df,hue="species",style="species")
```

##### conclusion 

1) Blue points can be easily separated from red and green by drawing a line.

2) But red and green data points cannot be easily separated.

3) Using sepal_length and sepal_width features and also petal_length and petal_width features, we can distinguish Setosa flowers from others.

4) Separating Versicolor from Viginica is much harder as they have considerable overlap.

### Pair plot

A pairs plot allows us to see both distribution of single variables and relationships between two variables.

```python
sns.pairplot(data = df,hue = "species")
```

##### conclusion 

1) The diagonal plot which showcases the histogram. The histogram allows us to see the PDF/Probability distribution of a single variable

2) Upper triangle and lower triangle which shows us the scatter plot.

3) The scatter plots show us the relationship between the features. These upper and lower triangles are the mirror image of each other.

4) petal length and petal width are the most useful features to identify various flower types.

5) While Setosa can be easily identified (linearly separable), virginica and Versicolor have some overlap (almost linearly separable).

### Histogram 

Histogram basically represents how many points exist for each value on the x-axis.

```python
# for petal length
sns.FacetGrid(df,hue="species",height = 6).map(sns.distplot,"petal_length").add_legend()
```

```python
# petal width
sns.FacetGrid(df,hue="species",height = 6).map(sns.distplot,"petal_width").add_legend()
```

As we compare petal length and petal width then petal length the setosa are better separable than using patel length.

```python
# for sepal length
sns.FacetGrid(df,hue="species",height = 6).map(sns.distplot,"sepal_length").add_legend()
```

```python
# for sepal_width 
sns.FacetGrid(df,hue="species",height = 6).map(sns.distplot,"sepal_width").add_legend()
```

##### conclusion 

1) using sepal_length and sepal_width we canâ€™t separate any class because all of them are overlapped

2) From above observations we can say that Sepal length and sepal width is worst than Petal length and Petal width.
